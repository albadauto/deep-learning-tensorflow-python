```python
X_train = training[:,1:] / 255
y_train = training[:, 0]
```

```python
X_test = testing[:,1:] / 255
y_test = testing[:, 0]
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train[0]
```

```python
#Para redes neurais convolucionais, o padrÃ£o Ã© a rede neural receber os dados no padrÃ£o (bat(quantos registros), Altura, largura, canal)
#Exemplo abaixo (60000, 28, 28, 1)
X_train = X_train.reshape(X_train.shape[0], *(28,28,1))
X_test = X_test.reshape(X_test.shape[0], *(28,28,1))
```

```python
X_train.shape
```

```python
#Treinamento
from tensorflow.keras import layers, models
cnn = models.Sequential()

cnn.add(layers.Conv2D(32, (3,3), activation = 'relu', input_shape=(28,28,1)))
cnn.add(layers.MaxPooling2D(2,2))

cnn.add(layers.Conv2D(64, (3,3), activation = 'relu'))
cnn.add(layers.MaxPooling2D(2,2))

cnn.add(layers.Conv2D(64, (3,3), activation = 'relu'))

cnn.add(layers.Flatten())

cnn.add(layers.Dense(64, activation = 'relu'))
cnn.add(layers.Dense(10, activation = 'softmax'))

cnn.summary()
```

```python
cnn.compile(loss = 'sparse_categorical_crossentropy', optimizer='Adam', metrics = ['accuracy'])
```

```python
epochs = 150
history = cnn.fit(X_train, y_train, batch_size = 512, epochs = epochs)
```

```python
evaluation = cnn.evaluate(X_test, y_test)
```

```python
evaluation
```

```python
predicted_classes = np.argmax(cnn.predict(X_test), axis=1)
```

```python
W_grid = 5
L_grid = 5
fig, axes = plt.subplots(L_grid, W_grid, figsize = (12,12))
axes = axes.ravel()

for i in np.arange(0, W_grid * L_grid):
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title("Prediction ckass = {:0.1f} \n True class = {:0.1f}".format(predicted_classes[i], y_test[i]), fontsize = 8)
    axes[i].axis('off')
plt.subplots_adjust(hspace=0.4)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot=True)
```
